```python
import os

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def set_color(number):
    if number == -1:
        return 'red'
    elif number == 0:
        return 'blue'
    elif number == 1:
        return 'green'
    elif number == 2:
        return 'orange'
```

```python
df = pd.read_csv('alabama.csv')
df.columns = ['index', 'a','b','c','d','e','f','g','h']

df['color'] = df['g'].apply(set_color)
df
```

```python
fig = plt.figure()
ax = plt.axes()

ax.scatter(df['index'], df['c'], c=df['color'])
plt.savefig('test.png')
```

```python
class PlotResults(object):

    def __init__(self, base_path):
        self._base_path = base_path

    def construct_df(self, filename):
        df = pd.read_csv(os.path.join(self._base_path, filename))
        df['color'] = df['state'].apply(set_color)
        return df

    @staticmethod
    def plot_df(df, log=False):
        fig = plt.figure()
        ax = plt.axes()
        plt.title('Log-Loss of Federated DPSGD')
        plt.xlabel('Step')
        plt.ylabel('Log-Loss')
        y = df['batch_loss'] if not log else np.log(df['batch_loss'])
        ax.scatter(df.index, y, c=df['color'])

    @staticmethod
    def plot_compound(dfs_list, log=False):
        fig = plt.figure()
        ax = plt.axes()
        plt.title('Log-Loss of Federated DPSGD')
        plt.xlabel('Step')
        if log:
            plt.ylabel('Log(Log-Loss)')
        else:
            plt.ylabel('Log-Loss')
        for df in dfs_list:
            y = df['batch_loss'] if not log else np.log(df['batch_loss'])
            ax.scatter(df.index, y, c=df['color'])

plotter = PlotResults('2021-03-06-05:55:36.558906')
df1 = plotter.construct_df('burn_in_alabama_1.csv')
df2 = plotter.construct_df('burn_in_alabama_2.csv')
df3 = plotter.construct_df('train_on_vt_3.csv')
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df = df.drop(0)
df = df.drop(1)
df = df.drop(2)
plotter.plot_df(df)
```

```python
df_1_3 = pd.concat([df1, df3], ignore_index=True)
df_1_3 = df_1_3.drop(0)
df_1_3 = df_1_3.drop(1)
df_1_3 = df_1_3.drop(2)
plotter.plot_df(df_1_3[:100])

```

```python
plotter.plot_compound([df_1_3[:97], df], log=True)
plt.savefig('medicare_federation_example_2021-03-06-05:55:36.558906-log.png', dpi=500)
```
