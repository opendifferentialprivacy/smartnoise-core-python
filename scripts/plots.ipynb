```python
def convert_int_to_statename(state_int):
    d = {-1: 'al', 0: 'al', 1: 'vt'}
    return d[state_int]

class PlotResults(object):

    def __init__(self, base_path):
        self._base_path = base_path
        self.dfs = []

    @staticmethod
    def set_color_from_statename(state):
        if state == 'al':
            return 'red'
        elif state == 'vt':
            return 'orange'
        elif state == 'ct':
            return 'green'
        elif state == 'ma':
            return 'blue'
        elif state == 'ri':
            return 'purple'
        elif state == 'nh':
            return 'gray'
        return 'black'

    def construct_df(self, filename, color=None):
        df = pd.read_csv(os.path.join(self._base_path, filename))
        if df.empty:
            return
        if color:
            df['color'] = color
        else:
            # df['state_name'] = df['state'].apply(convert_int_to_statename)
            df['color'] = df['state'].apply(self.set_color_from_statename)
        return df

    def generate_df_plot_inputs(self):
        """
        Generate list of dicts, each with state name and a DF.
        The result gets passed to plot_compound.
        :return:
        """
        files = os.listdir(self._base_path)
        csvs = [f for f in files if '.csv' in f]
        sorted_csvs = sorted(csvs, key=lambda x: int(x.split('_')[-1].split('.')[0]))
        states = [x.split('_')[-2] for x in sorted_csvs]

        self.dfs = []
        for state, csv in zip(states, sorted_csvs):
            df = plotter.construct_df(csv)
            if df is not None:
                self.dfs.append({'state': state, 'df': df})

    @staticmethod
    def plot_df(df, log=False):
        fig = plt.figure()
        ax = plt.axes()
        plt.title('Log-Loss of Federated DPSGD')
        plt.xlabel('Step')
        plt.ylabel('Log-Loss')
        y = df['batch_loss'] if not log else np.log(df['batch_loss'])
        ax.scatter(df.index, y, c=df['color'])

    def plot_compound(self, log=False):
        fig = plt.figure()
        ax = plt.axes()
        plt.title('Log-Loss of Federated DPSGD')
        plt.xlabel('Step')
        if log:
            plt.ylabel('Log(Log-Loss)')
        else:
            plt.ylabel('Log-Loss')

        # burn in length
        al_starting_point = self.dfs[0]['df'].shape[0]
        other_starting_point = self.dfs[0]['df'].shape[0]

        labels = [x['state'] for x in self.dfs]
        labels_used = []
        for i, df_dict in enumerate(self.dfs):
            df = df_dict['df']
            y = df['batch_loss'] if not log else np.log(df['batch_loss'])
            if i == 0:
                ax.scatter(df.index, y, c=df['color'])
            else:
                if df_dict['state'] == 'al':
                    df.index += al_starting_point
                    al_starting_point += df.shape[0]
                else:
                    df.index += other_starting_point
                    other_starting_point += df.shape[0]
                if labels[i] not in labels_used:
                    ax.scatter(df.index, y, c=df['color'], label=labels[i])
                    labels_used.append(labels[i])
                else:
                    ax.scatter(df.index, y, c=df['color'])
        ax.legend()

    def calculate_loss_difference(self):
        last_al_df = list(filter(lambda x: x['state'] == 'al', self.dfs))[-1]
        last_federated_df = self.dfs[-1]
        return last_al_df['df'].batch_loss.iloc[-1] - last_federated_df['df'].batch_loss.iloc[-1]
```

```python
base_dir = '2021-03-23-03:32:19.963211'
plotter = PlotResults(base_dir)
plotter.generate_df_plot_inputs()
plotter.plot_compound(log=True)
loss_difference = plotter.calculate_loss_difference()
print(f'Difference between AL and Federation: {loss_difference}')
```
